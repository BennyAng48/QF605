```python
def g_0(K):
    return K

def g_1(K):
    return 1.0

def g_2(K):
    return 0.0
```

### IRR and its derivatives

The **Instantaneous Annuity Factor** and its first two derivatives with respect to strike \( K \) are given by:

#### IRR(K)

$$
\text{IRR}(K) = \frac{1}{K} \left(1 - \frac{1}{\left(1 + \frac{K}{m}\right)^{m \cdot \text{tenor}}} \right)
$$

#### IRRâ€²(K)

$$
\text{IRR}'(K) = -\frac{1}{K} \cdot \text{IRR}(K) + \frac{1}{K \cdot m} \cdot \frac{m \cdot \text{tenor}}{\left(1 + \frac{K}{m} \right)^{m \cdot \text{tenor} + 1}}
$$

#### IRRâ€³(K)

$$
\text{IRR}''(K) = -\frac{2}{K} \cdot \text{IRR}'(K) - \frac{1}{K \cdot m^2} \cdot \frac{m \cdot \text{tenor} \cdot (m \cdot \text{tenor} + 1)}{\left(1 + \frac{K}{m} \right)^{m \cdot \text{tenor} + 2}}
$$


```python
def IRR_0(K, m, tenor):
    # Implementation of IRR(K) function
    value = 1/K * ( 1.0 - 1/(1 + K/m)**(tenor*m) )
    return value

def IRR_1(K, m, tenor):
    # Implementation of IRR'(K) function (1st derivative)
    firstDerivative = -1/K*IRR_0(K, m, tenor) + 1/(K*m)*tenor*m/(1+K/m)**(tenor*m+1)
    return firstDerivative

def IRR_2(K, m, tenor):
    # Implementation of IRR''(K) function (2nd derivative)
    secondDerivative = -2/K*IRR_1(K, m, tenor) - 1/(K*m*m)*(tenor*m)*(tenor*m+1)/(1+K/m)**(tenor*m+2)
    return secondDerivative
```

### Convexity-Adjusted Function \( h(K) \) and its Derivatives

Let \( h(K) = \frac{g(K)}{\text{IRR}(K)} \), where \( g(K) = K \).  
Then the first and second derivatives of \( h(K) \) are:

#### \( h(K) \)

$$
h(K) = \frac{g_0(K)}{\text{IRR}(K)} = \frac{K}{\text{IRR}(K)}
$$

#### \( h'(K) \)

$$
h'(K) = \frac{\text{IRR}(K) \cdot g_1(K) - g_0(K) \cdot \text{IRR}'(K)}{\text{IRR}(K)^2}
$$

#### \( h''(K) \)

$$
h''(K) = \frac{g_2(K) \cdot \text{IRR}(K) - g_0(K) \cdot \text{IRR}''(K) - 2 \cdot g_1(K) \cdot \text{IRR}'(K)}{\text{IRR}(K)^2}
+ \frac{2 \cdot g_0(K) \cdot \left(\text{IRR}'(K)\right)^2}{\text{IRR}(K)^3}
$$


```python
def h_0(K, m, tenor):
    # implementation of h(K)
    value = g_0(K) / IRR_0(K, m, tenor)
    return value

def h_1(K, m, tenor):
    # implementation of h'(K) (1st derivative)
    firstDerivative = (IRR_0(K, m, tenor)*g_1(K) - g_0(K)*IRR_1(K, m, tenor)) / IRR_0(K, m, tenor)**2
    return firstDerivative

def h_2(K, m, tenor):
    # implementation of h''(K) (2nd derivative)
    secondDerivative = ((IRR_0(K, m, tenor)*g_2(K) - IRR_2(K, m, tenor)*g_0(K) - 2.0*IRR_1(K, m, tenor)*g_1(K))/IRR_0(K, m, tenor)**2 
                        + 2.0*IRR_1(K, m, tenor)**2*g_0(K)/IRR_0(K, m, tenor)**3)
    return secondDerivative
```

```python
## Linear interpolate Libor DF
def LIBORDF(T, 
            df = Discount_Factors):
    x = list(df['Years'])
    y = list(df['IRS_Discount_Factor'])
    return np.interp(T, x, y)

## Linear interpolate OIS DF
def OISDF(T, 
          df = Discount_Factors):
    x = list(df['Years'])
    y = list(df['OIS_Discount_Factor'])
    return np.interp(T, x, y)

def Forward_LIBOR(m, 
                  D1, 
                  D2):
    return m * ((D1 - D2) / D2)
```

```python
## Calculate swap rate
def Forward_Swap_Rate(expiry, 
                      tenor, 
                      m):
    """
    Computes the forward par swap rate starting at `expiry`, with `tenor` years,
    and `m` fixed payments per year (e.g., 2 for semiannual).
    """
    floated, fix = 0, 0
    for i in range(1, int(tenor * m)+1):
        end = expiry + i/m
        floated += OISDF(end) * Forward_LIBOR(m, LIBORDF(end-1/m), LIBORDF(end)) * 1/m
        fix += OISDF(expiry + i/m)/m
    return floated/fix
```

```python
## Calculate the SABR implied volatitlity
def SABR(F, 
         K, 
         T, 
         alpha, 
         beta, 
         rho, 
         nu):
    
    """
    Compute the SABR implied volatility.

    Parameters:
    F: Forward price
    K: Strike price
    T: Time to maturity
    alpha: Volatility of the underlying asset
    beta: SABR exponent
    rho: Correlation between the asset price and its volatility
    nu: Volatility of volatility

    Returns:
    sabrsigma: SABR implied volatility
    """
    
    X = K
    if F == K:
        numer1 = (((1 - beta)**2)/24)*alpha*alpha/(F**(2 - 2*beta))
        numer2 = 0.25*rho*beta*nu*alpha/(F**(1 - beta))
        numer3 = ((2 - 3*rho*rho)/24)*nu*nu
        VolAtm = alpha*(1 + (numer1 + numer2 + numer3)*T)/(F**(1-beta))
        sabrsigma = VolAtm
    else:
        z = (nu/alpha)*((F*X)**(0.5*(1-beta)))*np.log(F/X)
        zhi = np.log((((1 - 2*rho*z + z*z)**0.5) + z - rho)/(1 - rho))
        numer1 = (((1 - beta)**2)/24)*((alpha*alpha)/((F*X)**(1 - beta)))
        numer2 = 0.25*rho*beta*nu*alpha/((F*X)**((1 - beta)/2))
        numer3 = ((2 - 3*rho*rho)/24)*nu*nu
        numer = alpha*(1 + (numer1 + numer2 + numer3)*T)*z
        denom1 = ((1 - beta)**2/24)*(np.log(F/X))**2
        denom2 = (((1 - beta)**4)/1920)*((np.log(F/X))**4)
        denom = ((F*X)**((1 - beta)/2))*(1 + denom1 + denom2)*zhi
        sabrsigma = numer/denom
        
    return sabrsigma
```

```python
# Linear interpolate SABR parameters
def ParameterInterp (dataframe, 
                     tenor, 
                     expiry):
    
    """
    Linearly interpolate SABR model parameters based on tenor and expiry.

    Args:
    dataframe: DataFrame containing SABR parameters for different tenors
    tenor: The tenor to interpolate (e.g., 2, 3, 5, 10)
    expiry: The expiry to interpolate (e.g., 1, 2, 5, 10)

    Returns:
    Interpolated value of SABR parameter for the given tenor and expiry.
    """
    
    # Step 1: Interpolate parameters for the given tenor across the known tenors [1, 5, 10]
    linear_interp = []
    new_paras = []
    
    for i in [1,2,3,5,10]:      # These represent the available tenors
        linear_interp.append(np.interp(tenor, [1, 5, 10], dataframe[i]))
    
    for i in range(5):
        new_paras.append(linear_interp[i])
    
    # Step 2: Interpolate the results for the given expiry
    result = np.interp(expiry, [1, 2, 3, 5, 10], new_paras)
    
    return float(result)
```

```python
## Black 76 pricing formula
def black76_payer(F, K, t, sigma):
    d1 = (np.log(F/K)+(1/2)*(sigma**2)*t)/(sigma*np.sqrt(t))
    d2 = d1 - sigma * np.sqrt(t)
    price = F*norm.cdf(d1) - K*norm.cdf(d2)
    return price

def black76_receiver(F, K, t, sigma):
    d1 = (np.log(F/K)+1/2*(sigma**2)*t)/(sigma*np.sqrt(t))
    d2 = d1 - sigma * np.sqrt(t)
    price = K*norm.cdf(-d2) - F*norm.cdf(-d1)
    return price
```

```python
## IRR-Settled payer and recevier swaption integrant at time T
def Rec_Integral(F, K, m, tenor, t, sigma):
    """
    Calculates the integral for the receiver swaption at time T.
    
    Parameters:
    F (float): Forward price.
    K (float): Strike price.
    m (int): Frequency of payments.
    tenor (float): The tenor of the swap (e.g., 5 years).
    t (float): Time to maturity (in years).
    sigma (float): Volatility of the underlying swaption.
    
    Returns:
    float: The receiver swaption integral value.
    """
    return h_2(K, m, tenor) * black76_receiver(F, K, t, sigma)

def Pay_Integral(F, K, m, tenor, t, sigma):
    """
    Calculates the integral for the payer swaption at time T.
    
    Parameters:
    F (float): Forward price.
    K (float): Strike price.
    m (int): Frequency of payments.
    tenor (float): The tenor of the swap (e.g., 5 years).
    t (float): Time to maturity (in years).
    sigma (float): Volatility of the underlying swaption.
    
    Returns:
    float: The payer swaption integral value.
    """
    return h_2(K, m, tenor) * black76_payer(F, K, t, sigma)
```

```python
## Calculate the CMS rate
def CMS(m, tenor, expiry):
    """
    Calculate the CMS rate based on the forward swap rate, 
    and integrals of receiver and payer swaption prices.

    Parameters:
    m (int): Payment frequency (e.g., semi-annual).
    tenor (float): The tenor of the swap (e.g., 5 years).
    expiry (float): Time to maturity (e.g., 1 year).
    
    Returns:
    float: The CMS rate.
    """
    
    # Get forward swap rate F
    F = Forward_Swap_Rate(expiry, tenor, m)
    
    # print(f"Forward Swap Rate is {F}")
    # Interpolate the SABR parameters: alpha, rho, and nu
    alpha = ParameterInterp(Alpha, tenor, expiry)
    rho = ParameterInterp(Rho, tenor ,expiry)
    nu = ParameterInterp(Nu, tenor ,expiry)
    
    Discount_factor = (OISDF(expiry))
    # Integrate over the receiver swaption strike range (0 to F) and payer swaption strike range (F to 0.1)
    CMS = \
            (
                Discount_factor*F 
                + quad(lambda x: Rec_Integral(F, x, m, tenor ,expiry, SABR(F,x,expiry,alpha,0.9,rho,nu)), 0, F)[0]
                + quad(lambda x: Pay_Integral(F, x, m, tenor ,expiry, SABR(F,x,expiry,alpha,0.9,rho,nu)), F, 0.1)[0]  # The upper range of integration is 0.2
                )
            
    return CMS
```

```python
## Calculate the present value of the CMS leg
def PV_CMS(m, tenor, expiry):
    n = int(expiry*m)           # Number of coupon payments
    pv = 0
    for i in range(1,n+1):      # For each coupon payment
        ti = i/m                # Get the time (in years) it is supposed to be paid out
        pv += (OISDF(ti)/m) * CMS(m, tenor, expiry)     # Calculate its present value by discounting by OIS discount factor
    return pv
```

```python
cms_start_time = [1, 5, 10]
cms_duration = [1, 2, 3, 5, 10]
cms_forward_swap = [[x, y] for x in cms_start_time for y in cms_duration]
cms_forward_swap
```

```python
cms_rates = []
for forward_swap in cms_forward_swap:
    expiry, tenor = forward_swap
    discount_factor = Discount_Factors.loc[tenor, "OIS_Discount_Factor"]
    swap_rate = CMS(m = 2, 
                    tenor = tenor, 
                    expiry = expiry)
    
    cms_rates.append([expiry, tenor, swap_rate])

cms_rates_df = pd.DataFrame(
    cms_rates,
    columns=[
        "Expiry",
        "Tenor",
        "CMS Rate",
    ],
)
```
